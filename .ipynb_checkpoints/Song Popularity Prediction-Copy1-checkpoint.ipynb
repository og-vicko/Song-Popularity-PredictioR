Song Popularity Prediction

Music streaming platforms have vast collections of songs, each with a varying degree of popularity among listeners. Understanding what makes a song popular is valuable for music producers, artists, and platforms curating content for their audiences. 

The ability to predict song popularity will enable music platforms to optimize their recommendation systems, improve user engagement, and assist music producers and artists in creating content that resonates with the audience.

The primary goal of this machine learning task is to predict the popularity of songs based on a range of features extracted from the music data. The dataset includes 40,000 entries with 15 distinct features such as song duration, acoustics, danceability, energy, instrumentals, key, liveness, loudness, audio mode, speechiness, tempo, time signature, and audio valence.

Build a machine learning model that provides reliable predictions of song popularity, which can be used to make data-driven decisions in music production and distribution.


Features

1. id: Unique identifier for each song.
2. song_duration_ms: Duration of the song in milliseconds.
3. acousticness: A measure of the acoustic quality of the song.
4. danceability: A measure of how suitable a song is for dancing.
5. energy: A measure of the intensity and activity of the song.
6. instrumentalness: Indicates the song's likelihood of being instrumental.
7. key: The key the song is in.
8. liveness: Detects the presence of an audience in the recording.
9. loudness: The overall loudness of a track in decibels.
10. audio_mode: Modality of the music (major or minor).
11. speechiness: Measures the presence of spoken words in a song.
12. tempo: The overall estimated tempo of a song.
13. time_signature: An estimated overall time signature of a song.
14. audio_valence: The musical positiveness conveyed by a song.
15. song_popularity: A score indicating the popularity of the song, which will be the target variable for our prediction model.

#### Handling Outliers

```python
fig, axes = plt.subplots(7, 2, figsize = (20, 20))
axes = axes.flatten()

for i, col in enumerate(train_.columns):
    if col != 'song_popularity':
        sns.boxplot(x = train_[col], ax = axes[i])
        axes[i].set_title(f'BoxPlot showing outliers in {col}')

plt.tight_layout()
plt.show()
```

The subplot above shows the distribution of each variable in the dataset. From this, we could also observe that a couple of these columns have outliers in them. Considering the kind of data we are working with and the problem we are trying to solve, it is important that these values are handled carefully. The variables with outliers includes; `Song duration ms`, `Energy`, `Instrumentals`, `Liveness`, `Loudness`,`Speechiness` and `Tempo`. Some have just a few, while others have multiple. To handle these outliers, we would be replacing the outliers with the lower and upper bounds of each variable (`Winsorization`). Further analysis would be carried out to validate the outlier status. 

```python
outliers_df = pd.DataFrame()

for feature in train_.columns:
    if feature != 'song_popularity':
        Q1 = train_[feature].quantile(0.25)
        Q3 = train_[feature].quantile(0.75)
        IQR = Q3 - Q1
        
        # Define the upper and lower bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify outliers using boolean indexing
        # outliers_lower = train_[feature] < lower_bound
        # outliers_upper = train_[feature] > upper_bound
        
        # # Concatenate outliers to the outliers_df DataFramee
        # outliers_df = pd.concat([outliers_df, train_[outliers_lower | outliers_upper]])
        # train_
        
        # Replace outliers with the upper/lower bound
        train_[feature] = train_[feature].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))
# outliers_df
```

#### Creating functions to be used.

```python
metrics_df = pd.DataFrame(columns = ['model', 'train_accuracy','val_accuracy', 'precision', 'recall', 'f1', 'logloss', 'roc'])

y_pred, y_pred_proba = 0, 0 

# Plot confusion matrix
def plot_confusion_matrix(classifier,y_val, ypred):
    plt.style.use('rose-pine-moon')
    sns.heatmap(confusion_matrix(y_val, ypred, normalize='true'), annot = True, cbar=False)
    plt.title(f'{classifier} Confusion Matrix')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.savefig(f'{classifier}_cf_matrix.png')
    plt.show();

# plot precision_recall_curve
def plot_precision_recall_curve(classifier,y_val, y_pred_proba):
    precision, recall, thresholds = precision_recall_curve(y_val, y_pred_proba)
    
    auc_pr = auc(recall, precision)
    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, color='blue', label=f'Precision-Recall Curve (AUC-PR = {auc_pr:.2f})')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'{classifier} Precision-Recall Curve')
    plt.legend()
    plt.grid(True)
    plt.savefig(f'{classifier}_precision_recall_curve.png')    
    plt.show()

# PLot roc_curve
def plot_roc_curve(classifier,y_val, y_pred_proba):
    with plt.style.context('rose-pine-moon'):
        fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)
        roc = roc_auc_score(y_val, y_pred_proba)
        plt.figure(figsize=(8, 4))
        plt.plot([0, 1], [0, 1], 'y--', label = 'Random Guess')
        plt.plot(fpr, tpr, lw=2, marker='.', color='blue', label='ROC Curve Area')
        plt.annotate(f"AUC = {roc:.2f}", xy=(0.7, 0.5), fontsize=12)
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'{classifier} ROC Curve')
        plt.legend(loc='best')
        plt.savefig(f'{classifier}_roc_curve.png')
        plt.show()

# Compute evaluation metrics
def compute_eval_metrics(y_val, ypred, ypred_train, ytrain=None, avg='weighted'):
    accuracy = accuracy_score(y_val, ypred)
    precision = precision_score(y_val, ypred, average=avg)
    recall = recall_score(y_val, ypred, average=avg)
    f1 = f1_score(y_val, ypred, average=avg)
    logloss = log_loss(y_val, ypred)
    roc = roc_auc_score(y_val, y_pred_proba)
     
    if ytrain is not None:
        accu_train = accuracy_score(ytrain, ypred_train) 
        
        return accu_train, accuracy, precision, recall, f1, logloss, roc
    


# Fit model, predict and call the eval metric funcs and plot funcs 
def train_predict_plot(model,classifier, xtrain, ytrain, xval = X_val, y_val = y_val):

    model.fit(xtrain, ytrain)
    
    global y_pred, y_pred_proba  

    y_pred_train = model.predict(xtrain)
    y_pred = model.predict(xval)
    y_pred_proba = model.predict_proba(xval)[:, 1]    
    
    train_accu, val_accu, precision, recall, f1, logloss, roc = compute_eval_metrics(y_val, y_pred, y_pred_train, y_train)
    metrics_df.loc[len(metrics_df.index)] = (classifier,train_accu,val_accu, precision, recall, f1, logloss, roc)
    
    print(f'Train Accuracy: {train_accu:.2f} | Val Accuracy: {val_accu:.2f} | Precision: {precision:.2f}', 
          f'| Recall: {recall:.2f} | F1-Score: {f1:.2f} | Log_Loss {logloss:.2f} | ROC-Auc-Score: {roc:.2f}')

    plot_confusion_matrix(classifier, y_val, y_pred)
    plot_precision_recall_curve(classifier, y_val, y_pred_proba)
    plot_roc_curve(classifier, y_val, y_pred_proba)

    return {'Train Accuracy': round(train_accu, 2),
            'Test Accuracy': round(val_accu, 2),
            'Precision': round(precision, 2), 
            'Recall': round(recall, 2), 
            'F1-Score': round(f1, 2),
            'Log_loss': round(logloss,2),
            'Roc_Auc_Score': round(roc, 2)}
```

```python
# Model to track experiment
def track_experiment(experiment_name, run_name, eval_metrics, model, cf_path=None, prc_path=None, roc_path=None, params=None, server=None):
    
    import mlflow
    import os

    # MLflow tracking server URI
    if server is not None:
        mlflow.set_tracking_uri(uri="http://localhost:5000")

    # set the Mlflow experiment
    mlflow.set_experiment(experiment_name)
    
    with mlflow.start_run():

        # Log hyperparameters
        if params is not None:
            mlflow.log_params(params)
        
        # Log metrics and model     
        mlflow.log_metrics(eval_metrics)
        mlflow.sklearn.log_model(model, "model")

        # check if image exists and log it if it does not  
        if cf_path is not None and os.path.exists(cf_path):
            mlflow.log_artifact(cf_path, 'confusion_matrix')
        else:
            print(f"Confusion Matrix file not found: {cf_path}")

        if prc_path is not None and os.path.exists(prc_path):
            mlflow.log_artifact(prc_path, 'precision_recall_curve')
        else:
            print(f"Precision-Recall Curve file not found: {prc_path}")

        if roc_path is not None and os.path.exists(roc_path):
            mlflow.log_artifact(roc_path, 'roc_auc_curve')
        else:
            print(f"ROC Curve file not found: {roc_path}")
        
        mlflow.set_tag("tag1", "Song Popularity Prediction")
        mlflow.set_tags({"tag2": f'{model}', "tag3": "Binary Classification"})
        
    print(f'Run Name: {run_name}\nExperiment Name: {experiment_name}\n')
    print(f'{run_name} has successfully been logged to {experiment_name} in MLFlow')
```

```python
import optuna
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Define the objective function to optimize
def objective(trial):
    # Define the hyperparameter search space
    n_estimators = trial.suggest_int('n_estimators', 50, 200)
    max_depth = trial.suggest_int('max_depth', 5, 30)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 10)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 4)
    bootstrap = trial.suggest_categorical('bootstrap', [True, False])

    # Create a Random Forest model with the suggested hyperparameters
    rf_model = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        bootstrap=bootstrap
    )

    # Use cross-validation to estimate model performance
    score = cross_val_score(rf_model, X_train_scaled, y_train, cv=5, scoring='roc_auc')

    # Return the average cross-validated ROC AUC as the objective value to minimize
    return np.mean(score)

# Create a study object and optimize the objective function
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

# Get the best hyperparameters from the study
best_params_optuna = study.best_params

# Create a Random Forest model with the best hyperparameters
# best_rf_model_optuna = RandomForestClassifier(**best_params_optuna)

# # Fit the model to the training data
# best_rf_model_optuna.fit(X_train_scaled, y_train)

# # Evaluate the best Random Forest model using ROC AUC on the test set
# roc_auc_optuna = best_rf_model_optuna.score(X_test, y_test)

# # Print the best hyperparameters and performance
# print("Best Hyperparameters:", best_params_optuna)
# print("ROC AUC on Test Set:", roc_auc_optuna)
```

##### Multilayer Paceptron

```python
class MultiLayerPerceptron(nn.Module):
    '''
       Parameters:
        - num_features (int): The number of input features.
        - num_classes (int): The number of output classes.

        Architecture:
            - Input Layer: Takes input features.
            - 1st Hidden Layer: Fully connected layer with 25 units followed by ReLU activation.
            - 2nd Hidden Layer: Fully connected layer with 15 units followed by ReLU activation.
            - Output Layer: Fully connected layer producing num_classes output units.
    
        Methods:
            - forward(x): Performs forward pass through the network and returns logits.
    '''
    def __init__(self, num_features, num_classes):
        """
        Initializes the MultiLayerPerceptron.

        Args:
            num_features (int): The number of input features.
            num_classes (int): The number of output classes.
        """
        super().__init__()
        self.all_layers = nn.Sequential(
            
            # 1st hidden layer
            nn.Linear(in_features = num_features, out_features=25),
            nn.ReLU(),

            # 2nd hidden layer
            nn.Linear(25,15),
            nn.ReLU(),

            # output Layer
            nn.Linear(15, num_classes)
        )

    def forward(self, x):
        '''
        Performs forward pass through the network.   
        '''
        logits = self.all_layers(x)
        return logits
```

```python
class MyDataset(Dataset):
    def __init__(self, x, y):
        self.features = torch.tensor(x.values, dtype = torch.float32)  
        self.labels = torch.tensor(y.values, dtype = torch.long) 

    def __getitem__(self, index):
       """
        Retrieves a single data sample and its label at the specified index.

        Args:
            index (int): Index of the data sample.

        Returns:
            torch.Tensor: Input features of the sample.
            torch.Tensor: Label of the sample.
        """        
        x = self.features[index]
        y = self.labels[index]
        return x, y

    def __len__(self):
        return self.labels.shape[0]

train_data = MyDataset(X_train_scaled, y_train)
val_data = MyDataset(X_val_scaled, y_val)
# test_data = MyDataset(X_test, y_test)

train_loader = DataLoader(
    dataset = train_data,
    shuffle = True,  
    batch_size = 32
)

val_loader = DataLoader(
    dataset = val_data,
    shuffle = False,  
    batch_size = 32
)
```

```python
# Creating a helper function

def compute_accuracy(model, dataloader):

    correct = 0
    total_examples = 0

    model = model.eval()

    for batch_idx, (features, labels) in enumerate(dataloader):

        with torch.no_grad():
            logits = model(features)

        predictions = torch.argmax(logits, dim = 1)

        compare = labels == predictions
        correct += torch.sum(compare)
        total_examples += len(compare)

    accuracy_score = correct / total_examples

    return accuracy_score
```

```python
model = MultiLayerPerceptron(13,2)

optimizer = torch.optim.SGD(params = model.parameters(), lr = 0.05)
criterion = nn.CrossEntropyLoss()
num_epochs = 10
```

initializing the MLP model with the `13 inputs` and `2` outputs. Stochastic Gradient Descent (SGD) is as ouotr optimizer and Cross-Entropy Loss as the los function

```python
# Training loop
for epoch in range(num_epochs):
    
    # Set the model to training mode
    model = model.train()

    # Iterate over batches in the training data
    for batch_idx, (features, labels) in enumerate(train_loader):
        # Forward pass
        logits = model(features)
        
        # Compute the Cross-Entropy loss
        loss = criterion(logits, labels)

        # Zero the gradients, backward pass, and optimizer step
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Logging for each batch
        print(f'Epoch: {epoch+1:03d}/{num_epochs:03d} | Batch {batch_idx+1:03d}/{len(train_loader):03d}',
              f' | Loss: {loss:.2f}')

    # Compute and log accuracy on training and validation sets
    train_acc = compute_accuracy(model, train_loader)
    val_acc = compute_accuracy(model, val_loader)
    print()
    print(f"At Epoch {epoch+1}: Train Acc {train_acc*100:.2f}% | Val Acc {val_acc*100:.2f}%")
    print('______'*10, '\n')
```
